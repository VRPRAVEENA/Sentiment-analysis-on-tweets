# YCSB Database Comparison between MongoDB and MYSQL

## Importing Files and creating dfs

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
files = []


# SQL file runs
files.append(('MYSQL Run A', 'outputMySQLRuna.txt'))
files.append(('MYSQL Run B', 'outputMySQLRunb.txt'))
files.append(('MYSQL Run C', 'outputMySQLRunc.txt'))
files.append(('MYSQL Run D', 'outputMySQLRund.txt'))
files.append(('MYSQL Run E', 'outputMySQLRune.txt'))
files.append(('MYSQL Run F', 'outputMySQLRunf.txt'))

# MongoDB file runs
files.append(('Mongo Run A', 'outputMongoRuna.txt'))
files.append(('Mongo Run B', 'outputMongoRunb.txt'))
files.append(('Mongo Run C', 'outputMongoRunc.txt'))
files.append(('Mongo Run D', 'outputMongoRund.txt'))
files.append(('Mongo Run E', 'outputMongoRune.txt'))
files.append(('Mongo Run F', 'outputMongoRunf.txt'))
```

```python
# creating an empty list that will store the ycsb run dataframes
# The first row needs to be removed as its not output data

results = []
cols = ['batchSize:-1']

for name, file in files:
    # read all csvs
    df = pd.read_csv(file, skiprows=1) # here i tell it to avoid row 1 as its junk
    df.columns=[name]
    df = df.T
    df = df.drop([x for x in cols if x in df.columns], axis =1)
    results.append(df)
    
df = pd.concat(results, axis = 0)
df.head()
```

```python
# Fix the multiple indexes in df
df.columns = df.columns.map('|'.join).str.strip('|')
```
