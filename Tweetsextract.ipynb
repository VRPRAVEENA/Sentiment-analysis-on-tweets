```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

from datetime import date, timedelta

def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)

start_date = date(2021, 11, 1)
end_date = date(2022, 11, 1)
end_date = start_date + timedelta(days = 365)
subject = "@taylorswift13"



for single_date in daterange(start_date, end_date):

    filepath = "./Tweets_" + single_date.strftime("%Y-%m-%d")+".csv"
    
    try:

        # Creating list to append tweet data to
        tweets_list2 = []
        
        currentDate = single_date.strftime("%Y-%m-%d")
        nextDate = (single_date+timedelta(days=1)).strftime("%Y-%m-%d")
        scrapeString = subject+'since:'+currentDate+' until:'+nextDate
        
        print(scrapeString)

        try:
        # Using TwitterSearchScraper to scrape data and append tweets to list
            for i,tweet in enumerate(sntwitter.TwitterSearchScraper(scrapeString).get_items()):
                if i>800:
                    break
                tweets_list2.append([tweet.id, tweet.date, tweet.content, tweet.hashtags,tweet.user.username])

        except:
            print("There was an error with tweet:" + i)        
            
        # Creating a dataframe from the tweets list above
        tweets_df2 = pd.DataFrame(tweets_list2, columns=['id', 'date', 'tweet', 'hashtags','username'])
        print(filepath)
        
        tweets_df2.to_csv(filepath)
    except:
        print("There was an error creating file:" + filepath)        
        
```
